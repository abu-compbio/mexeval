## This notebook is used to calculate subtype specific average TML for patient groups.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
c ='COADREAD'

# get strat file and prepare a dictionary of samples and subtypes
sub_file_path = '../mutex_data/molecular_strat_patient_list/'

filenames = {'BRCA': sub_file_path+'BRCA_molecularCBIO_patient_strat.txt',
            'COADREAD':sub_file_path + 'COADREAD_CMS_patient_strat_processed.txt'}

with open(filenames[c]) as f:
    dict_sub = {line.split()[0]:line.split()[1] for line in f.readlines()}
dict_sub
```

```python
#load binary matrix and TML from patient data
binary_matrix_infile = '../mutex_data/binary_matrices_all_genes_ep_mutation_filtered/{}_TML_binary_sm.txt'.format(c)

df = pd.read_csv(binary_matrix_infile, sep='\t', header=0, index_col=0)
dict_tml = df['y'].to_dict()
dict_tml
```

```python
d_main = {}

for p,tml in dict_tml.items():
    if len(p)== len(list(dict_sub)[0]):
        if p in dict_sub:
            subtype = dict_sub[p]
            if subtype not in d_main:
                d_main[subtype] = []

            d_main[subtype].append(tml)
            
    else:
        p=p[:12]
        if p in dict_sub:
            subtype = dict_sub[p]
            if subtype not in d_main:
                d_main[subtype] = []

            d_main[subtype].append(tml)
d_main
```

```python
import os
#write to file
outpath = '../results_main/subtype/average_TML/'
outfile = outpath + '/{}_average_tml_per_subtype.txt'.format(c)

if not os.path.exists(outpath):
    os.makedirs(outpath)

with open(outfile, 'w') as f:
    for k in sorted(d_main):
        f.write('{}\t{}\n'.format(k, np.sum(d_main[k])/float(len(d_main[k]))))
```
